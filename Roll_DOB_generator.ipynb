```
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import xlsxwriter
import pandas as pd
```

```
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
bot = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```
def login(unp,rowCounter):
    details = []
    time.sleep(0.5)
    try :
        username = bot.find_element_by_id('login')
    except :
        time.sleep(1)
        login(unp,rowCounter)
    
    password = bot.find_element_by_id('password')
    username.send_keys(unp)
    password.send_keys(unp)
    password.send_keys(Keys.RETURN)
    
    time.sleep(1)
    
    
    try:

        x = '//*[@id="featured_slide"]/table/tbody/'
        
        #roll
        details.append(bot.find_element_by_xpath(x + 'tr[3]/td[4]').text)
        #name
        details.append(bot.find_element_by_xpath(x + 'tr[4]/td[4]').text )
        #course
        details.append(bot.find_element_by_xpath(x + 'tr[1]/td[4]').text )
        #branch
        details.append(bot.find_element_by_xpath(x + 'tr[2]/td[4]').text )
        #DOB
        details.append(bot.find_element_by_xpath(x + 'tr[6]/td[4]').text)
    
    except:
        
        bot.get("http://www.griet.in/gcap/login.php?type=student")
        return None
    
    bot.find_element_by_xpath('//*[@id="header"]/div[2]/p/a[2]').click()
    return details
```

```
dataframe = pd.DataFrame(columns = ['rollno','Name','course','Branch','DOB']) 


rowCounter = 1
loopNo=1
count1=1
count3=0
alphCount = 'A'
while(loopNo <= 360):

    if(loopNo<10):
        unp = "18241A120" + str(count1)
        count1 += 1
    
    elif(loopNo < 100):
        unp = "18241A12" + str(count1)
        count1 += 1

    else:
        unp = "18241A12" + alphCount + str(count3)
        count3 = count3 + 1  
        if(count3 == 10):   
            alphCount = chr(ord(alphCount)  + 1)
            count3 = 0
    #print(unp,rowCounter)      
    details = login(unp,rowCounter)
    if details is not None : 
        print(details)
        a_series = pd.Series(details, index = dataframe.columns)
        dataframe = dataframe.append(a_series, ignore_index=True)
    else : pass
    rowCounter += 1
    loopNo += 1
```

```
dataframe['DDMMYYYY'] = [''.join(reversed(i.split('-'))) for i in dataframe['DOB']]
```
